```python
import pandas as pd

df = pd.read_csv(r"D:\Marketing-ROI-Analysis\data\marketing_campaign_dataset.csv")

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df['ctr'] = df['Clicks'] / df['Impressions']
df['conversions'] = df['Clicks'] * df['Conversion_Rate']
df['cac'] = df['Acquisition_Cost'] / df['conversions']
```

```python
df.groupby('Channel_Used')['ROI'].mean().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Channel_Used', y='ROI', data=df)
plt.xticks(rotation=45)
plt.show()
```
